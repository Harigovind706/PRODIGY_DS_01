Create a bar chart or histogram to visualize the distribution of
a categorical or continuous variable, such as the distribution of
ages or genders in a population.

This Excel file named WorldBank.xlsx includes various economic and social indicators for different countries, as provided by the World Bank. The data includes country names, region, income groups, and yearly records for statistics such as birth and death rates, GDP, internet usage, life expectancy, and population density. This allows for analysis of economic and social health on a country-by-country basis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'C:\\Users\\Hari\\Desktop\\Prodigy Infotech\\WorldBank.xlsx'
xls = pd.ExcelFile(file_path)

# Load the data from the "Development Indicators" sheet
df = pd.read_excel(file_path, sheet_name='Development Indicators')

# Select the variable "Life expectancy at birth (years)" for visualization
# Drop missing values to ensure the histogram is accurate
life_expectancy = df['Life expectancy at birth (years)'].dropna()

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(life_expectancy, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Life Expectancy at Birth (Years)')
plt.xlabel('Life Expectancy at Birth (Years)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'C:\\Users\\Hari\\Desktop\\Prodigy Infotech\\WorldBank.xlsx'
xls = pd.ExcelFile(file_path)

# Load the data from the "Development Indicators" sheet
df = pd.read_excel(file_path, sheet_name='Development Indicators')

# Filter the dataframe for the year 2018 and drop rows with missing GDP per capita values
gdp_2018 = df[df['Year'] == 2018][['Country Name', 'GDP per capita (USD)']].dropna()

# Sort the values by GDP per capita to make the bar chart more readable
gdp_2018 = gdp_2018.sort_values(by='GDP per capita (USD)', ascending=False).head(20)  # Top 20 countries

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.bar(gdp_2018['Country Name'], gdp_2018['GDP per capita (USD)'])
plt.title('Top 20 Countries by GDP per Capita in 2018')
plt.xlabel('Country')
plt.ylabel('GDP per Capita (USD)',)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
